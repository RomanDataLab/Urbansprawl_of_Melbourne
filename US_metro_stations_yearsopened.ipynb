```python
import pandas as pd
import wikipedia as wp
html = wp.page("List_of_railway_stations_in_Melbourne").html().encode("UTF-8")
try: 
    df = pd.read_html(html)[1]  # Try 2nd table first as most pages contain contents table first
except IndexError:
    df = pd.read_html(html)[0]
df.to_string()
df.head()
```

```python
df.info()
```

```python
dy = df.copy()  # Create a copy of the DataFrame

columns_to_remove = [1, 2, 4, 5]
# Dropping the specified columns
dy = dy.drop(dy.columns[columns_to_remove], axis=1)

# Define the new column names based on the indices
new_column_names = {0: 'name', 1: 'lines', 2: 'tariff_zone', 3: 'age', 4: 'loc', 5: 'notes'}

# Assign new column names directly
dy.columns = [new_column_names.get(idx, col) for idx, col in enumerate(dy.columns)]

dy.keys()
```

```python
dy.head()
```

```python
dy['tariff_zone'] = dy['tariff_zone'].str.split('[').str[0]
dy['age'] = dy['age'].str.split(' ').str[-1].str.split('[').str[0].astype(int)
dy['loc'] = dy['loc'].str.split('[').str[0]
dy.head()
```

```python
dy['notes'] = dy['notes'].str.split('[').str[0]
dy.head()
```

```python
import geopandas as gpd
import numpy as np
gdf = gpd.read_file(r'C:\WORK_IAAC\submissions\UrbanSense\UrbanSens_QGIS\VictoriaState\transport\PTV+station\PTV_METRO_TRAIN_STATION.shp')
gdf.head()
```

```python
gdf['loc'] = gdf['STOP_NAME'].str.extract(r'\((.*?)\)')
gdf['name'] = gdf['STOP_NAME'].str.split(' Railway').str[0]
gdf.head()
```

```python
# Count the number of coincidences
coincidences = gdf['name'].isin(dy['name']).sum()

# Calculate NaN values
nan_values = gdf['name'].isna().sum()

# Calculate non-coincidences
non_coincidences = gdf['name'].isin(dy['name']).value_counts()[False]

# Print the results
print("Number of coincidences:", coincidences)
print("Number of NaN values:", nan_values)
print("Number of non-coincidences:", non_coincidences)

# Get NaN and non-coincidences
nan_and_non_coincidences = gdf[~gdf['name'].isin(dy['name']) | gdf['name'].isna()]

nan_and_non_coincidences
```

```python
gdf.loc[gdf['name'] == 'Jolimont-MCG', 'name'] = 'Jolimont'
gdf.loc[gdf['name'] == 'Glenhuntly', 'name'] = 'Glen Huntly'
```

```python
not_in_gdf = dy[~dy['name'].isin(gdf['name'])]
not_in_gdf
```

```python
# Find rows containing 'Showgrounds'
showgrounds_rows = gdf[gdf['STOP_NAME'].str.contains('Showgrounds')]

# Find rows containing 'Flemington Racecourse'
flemington_rows = gdf[gdf['STOP_NAME'].str.contains('Flemington Racecourse')]

# Display the resulting rows
print("Rows containing 'Showgrounds':")
print(showgrounds_rows)

print("\nRows containing 'Flemington Racecourse':")
print(flemington_rows)
```

```python
from shapely.geometry import Point
# Specify the values for the new row
new_row = {
    'STOP_ID': int(gdf['STOP_ID'].max()) + 1,
    'STOP_NAME': 'Union',
    'name': 'Union',
    'LATITUDE': (gdf['LATITUDE'].iloc[145] + gdf['LATITUDE'].iloc[146]) / 2,
    'LONGITUDE': (gdf['LONGITUDE'].iloc[145] + gdf['LONGITUDE'].iloc[146]) / 2,
    'TICKETZONE': gdf['TICKETZONE'].iloc[145],  # Inherit from an existing row
    'ROUTEUSSP': gdf['ROUTEUSSP'].iloc[145],    # Inherit from an existing row
    'loc': gdf['loc'].iloc[145] + gdf['loc'].iloc[146]  # Concatenate strings from two rows
}

# Calculate the midpoint
point1 = gdf['geometry'].iloc[145]
point2 = gdf['geometry'].iloc[146]
midpoint = Point((point1.x + point2.x) / 2, (point1.y + point2.y) / 2)

# Add the 'geometry' column to the new row dictionary
new_row['geometry'] = midpoint

# Create the new row DataFrame
new_row_df = pd.DataFrame([new_row])

# Convert the new row DataFrame to a GeoDataFrame
new_row_gdf = gpd.GeoDataFrame(new_row_df, geometry='geometry')

# Append the new row to gdf
gdf = pd.concat([gdf, new_row_gdf], ignore_index=True)

gdf.info()
```

```python
dy.info()
```

```python
# Extract the desired columns from dy
dfx = dy[['name','age', 'loc', 'notes']]

# Merge dfx with gdf based on the 'name' column
merged_gdf = gdf.merge(dfx, on='name', how='left')
merged_gdf
```

```python
columns_to_remove = [-2]
merged_gdf = merged_gdf.drop(merged_gdf.columns[columns_to_remove], axis=1)
# Check if 'name' and 'notes' columns exist in merged_gdf
if 'name' in merged_gdf.columns and 'notes' in merged_gdf.columns:
    # Add 'closed' to rows with 'name' values 'Surrey Hills' and 'Mont Albert'
    merged_gdf.loc[merged_gdf['name'].isin(['Surrey Hills', 'Mont Albert']), 'notes'] += ' closed'
    
    # Add 'festival only' to rows with 'name' values 'Flemington Racecourse' and 'Showgrounds'
    merged_gdf.loc[merged_gdf['name'].isin(['Flemington Racecourse', 'Showgrounds']), 'notes'] += ' festival only'
merged_gdf
```
